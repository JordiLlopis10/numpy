## Arrays de tipo fijo en Python

Python ofrece varias opciones para almacenar datos en búferes de datos eficientes de tipo fijo. El módulo integrado `array` puede usarse para crear arrays densode un tipo uniforme:

````python
import array
lista_enteros = list(range(10))
arr_enteros = array.array('i', lista_enteros)
arr_enteros
````

Aquí `'i'` es un código de tipo que indica que el contenido son enteros (`int`)

Sin embargo, mucho más útil es el objeto `np.array` del paquete NumPy. Mientras que el objeto `array` de Python proporciona almacenamiento eficiente de datos basados en arrays, <span style="color:Gold">NumPy añade *operaciones* eficientes sobre esos datos</span>.

Y por eso lo usaremos en análisis de datos.

Vamos a crear arrays con NumPy.

Comenzaremos con la importación estándar de NumPy, bajo el alias `np`:

````python
import numpy as np
````

---
## Creación de Arrays a partir de Listas de Python

En primer lugar, podemos usar `np.array` para crear arrays a partir de listas de Python:

````python
# array de enteros:
np.array([1, 4, 2, 5, 6])
````

Recuerda que, a diferencia de las listas de Python, NumPy está limitado a arrays que contienen todos el mismo tipo. Si los tipos no coinciden, <span style="color:Gold">NumPy realizará un upgrade/promoción</span> de tipo (*upcast*) si es posible (aquí, los enteros se upgradean a punto flotante):

````python
# mezcla de entero y flotante --> se promueve todo a flotante
np.array([3.14, 4.2, 2, 3])
````

Si queremos establecer explícitamente el tipo de datos del array resultante, podemos usar el argumento `dtype`:

````python
# forzar tipo flotante de 32 bits
np.array([1, 2, 3, 4], dtype='float32')
````

<b>Nota:</b> Es clave familiarizarnos con el argumento dtype.

## Creación de arrays desde cero

Especialmente para arrays grandes, es más eficiente crear arrays desde cero utilizando las funciones de NumPy. Aquí hay varios ejemplos:

````python
# Crear un array de enteros de longitud 10 relleno de ceros
np.zeros(10, dtype=int)
````

````python
# Crear un array de punto flotante de 20 relleno de unos
np.ones(20, dtype=float)
````

````python
# Crear un array 3x5 relleno con el valor 3.14
np.full((3, 5), 3.14)
````

````python
# Crear un array relleno con una secuencia lineal
# Empezando en 0, terminando en 20, con paso de 2
# (similar a la función integrada range())
np.arange(0, 20, 2)
````

````python
# Crear un array de cinco valores entre 0 y 1 sepasrados por la misma distancia
np.linspace(0, 1, 5)
````

````python
# Crear un array de tam 5 de valores aleatorios con distribución uniforme de 0 a 1
np.random.random(5)
````

## Tipos de datos estándar de NumPy

Los arrays NumPy contienen valores de un único tipo, por lo que es importante conocer en detalle esos tipos y sus limitaciones. NumPy está desarrollado en C.

Los tipos de datos estándar de NumPy se listan en la siguiente tabla. Ten en cuenta que al construir un array pueden especificarse mediante una cadena de texto:

```python
np.zeros(10, dtype='int16')
```

O usando el objeto NumPy asociado:

```python
np.zeros(10, dtype=np.int16)
```

| Tipo de dato | Descripción |
|---|---|
| `bool_` | Booleano (True o False) almacenado como un byte |
| `int_` | Tipo entero por defecto (igual que `long` en C; normalmente `int64` o `int32`) |
| `intc` | Idéntico a `int` en C (normalmente `int32` o `int64`) |
| `intp` | Entero usado para indexación (igual que `ssize_t` en C; normalmente `int32` o `int64`) |
| `int8` | Byte (-128 a 127) |
| `int16` | Entero (-32768 a 32767) |
| `int32` | Entero (-2147483648 a 2147483647) |
| `int64` | Entero (-9223372036854775808 a 9223372036854775807) |
| `uint8` | Entero sin signo (0 a 255) |
| `uint16` | Entero sin signo (0 a 65535) |
| `uint32` | Entero sin signo (0 a 4294967295) |
| `uint64` | Entero sin signo (0 a 18446744073709551615) |
| `float_` | Abreviatura de `float64` |
| `float16` | Flotante de media precisión: bit de signo, 5 bits de exponente, 10 bits de mantisa |
| `float32` | Flotante de simple precisión: bit de signo, 8 bits de exponente, 23 bits de mantisa |
| `float64` | Flotante de doble precisión: bit de signo, 11 bits de exponente, 52 bits de mantisa |
| `complex_` | Abreviatura de `complex128` |
| `complex64` | Número complejo representado por dos flotantes de 32 bits |
| `complex128` | Número complejo representado por dos flotantes de 64 bits |
