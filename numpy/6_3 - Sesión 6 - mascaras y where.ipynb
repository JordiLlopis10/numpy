### Ejercicio 1 — Etiquetado condicional de acciones bursátiles

**Enunciado:**  
Tienes la variación diaria (%) de 10 acciones. Usa `np.where()` para crear un array de etiquetas:  
`'SUBE'` si la variación es positiva, `'BAJA'` si es negativa, y `'ESTABLE'` si es exactamente 0.

**Ten en cuenta:**
- `np.where(cond, valor_true, valor_false)` aplica una selección elemento a elemento.
- Para tres categorías puedes anidar dos `np.where()`: `np.where(cond1, A, np.where(cond2, B, C))`.


```python
variaciones = np.array([1.2, -0.8, 0.0, 3.5, -2.1, 0.0, 0.5, -1.4, 2.8, -0.3])

etiquetas = np.where(
    variaciones>0,
    'SUBE',
    np.where(variaciones<0,
             'BAJA',
             'ESTABLE')
    )

print(etiquetas)
```

### Ejercicio 2 — Bonificación salarial condicional

**Enunciado:**  
Una empresa calcula bonificaciones según rendimiento (0-100) y antigüedad (años).  
Usa `np.where()` para calcular el salario final:
- Rendimiento >= 80 **y** antigüedad >= 5 -> salario x 1.20  
- Rendimiento >= 80 **o** antigüedad >= 5 -> salario x 1.10  
- En otro caso -> salario x 1.00

**Ten en cuenta:**
- Calcula primero los multiplicadores con `np.where` anidado.
- Combina condiciones con `&` y `|` dentro de `np.where`.


```python
salarios     = np.array([30000, 45000, 28000, 60000, 35000, 52000, 40000])
rendimiento  = np.array([85,    72,    90,    65,    80,    88,    55   ])
antiguedad   = np.array([7,     3,     2,     10,    5,     6,     1    ])

mascara_top = (rendimiento>=80) & (antiguedad>=5) # x1.20 = incremento del 20%
mascara_mid = (rendimiento>=80) | (antiguedad>=5) # x1.10 = incremento del 10%

multiplicador = np.where(
    mascara_top,
    1.20,
    np.where(
        mascara_mid,
        1.10,
        1
    )
)
nuevos_salarios = salarios * multiplicador
print(nuevos_salarios)
```

### Ejercicio 3 — Pipeline completo de análisis de calidad

**Enunciado:**  
Una fábrica mide 3 parámetros de calidad en 8 lotes: temperatura, pureza y tiempo de ciclo.  
Un lote es **válido** solo si los 3 parámetros están dentro de rango.  
Usa **todo lo aprendido** (máscaras 2D, combinadas, reemplazo y `np.where`) para: identificar lotes válidos, puntuar cada lote (0-3) y crear un informe de estado.

**Rangos aceptables:** temperatura [180-205 °C], pureza >= 95%, ciclo [38-48 s]

**Ten en cuenta:**
- Crea una máscara 1D separada para cada parámetro (usando `lotes[:, col]`).
- Suma las máscaras (True=1) para obtener la puntuación por lote.
- Usa `np.where()` anidado para asignar: APROBADO / REVISION / RECHAZADO.


```python
# Columnas: temperatura (°C), pureza (%), tiempo_ciclo (s)
lotes = np.array([
    [185.0, 98.5, 42.0],
    [210.0, 95.0, 38.5],
    [175.0, 89.0, 45.0],
    [195.0, 99.0, 41.5],
    [220.0, 97.0, 55.0],
    [180.0, 92.0, 39.0],
    [190.0, 100.0, 43.0],
    [170.0, 88.0, 35.0]
])

mask_temp = np.where((lotes[:0] > 180) | (lotes[:0] < 205 ),True,False)
mask_pur = np.where((lotes[0:1] >= 95),True,False)
mask_ciclo = np.where((lotes[1:2] > 38) | (lotes[1:2] < 48 ),True,False)

```

**Variante extra:** Calcula para cada parámetro qué porcentaje de lotes lo superan, y añade una columna de 'desviación máxima' en los lotes rechazados.

```python
# CHULETA RESUMEN
# 

import numpy as np

a = np.array([1, 5, 3, 8, 2, 7])

# Máscara básica
mask = a > 4                        # [F T F T F T]

# Filtrar
print("Filtrado:   ", a[mask])      # [5 8 7]

# Combinar condiciones
print("AND:        ", a[(a > 2) & (a < 8)])
print("OR:         ", a[(a < 2) | (a > 6)])
print("NOT:        ", a[~(a > 4)])

# Índices de True
print("Indices:    ", np.where(mask)[0])

# Reemplazo in-place
b = a.copy()
b[mask] = 0
print("Reemplazo:  ", b)

# Selección condicional
print("np.where(): ", np.where(a > 4, a, 0))
print("np.where(): ", np.where(a > 4, 'alto', 'bajo'))
```
