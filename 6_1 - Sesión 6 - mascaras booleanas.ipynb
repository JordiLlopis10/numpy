### Ejercicio 1 — Temperaturas por encima del umbral

**Enunciado:**  
Una estación meteorológica registró las temperaturas diarias (°C) durante dos semanas.  
Filtra y devuelve solo las temperaturas que superan los 30°C (ola de calor).

**Ten en cuenta:**
- Una máscara booleana es un array de `True/False` del mismo tamaño.
- Puedes crear una con `array > valor`.
- Aplica la máscara al array original usando corchetes: `array[mascara]`.


```python
temperaturas = np.array([22.1, 31.5, 28.4, 35.2, 19.8, 33.0, 27.6,
                         30.1, 29.9, 36.4, 21.3, 32.8, 18.5, 34.7])
mascara_30 = temperaturas > 30
oladecalor = temperaturas[mascara_30]
print(oladecalor)
```

**Variante extra:** Calcula también el promedio de temperaturas durante los días de ola de calor.

### Ejercicio 2 — Notas de estudiantes aprobados

**Enunciado:**  
Un profesor tiene las notas finales de 12 estudiantes (escala 0-10).  
Extrae las notas de quienes aprobaron (nota >= 5) y cuenta cuántos suspendieron.

**Ten en cuenta:**
- `~mascara` invierte una máscara booleana (`True` <-> `False`).
- `.sum()` sobre una máscara cuenta los `True`.


```python
notas = np.array([7.5, 3.2, 9.1, 4.8, 6.0, 2.5, 8.3, 5.0, 1.9, 7.7, 4.1, 6.5])

mascara_nota5 = notas >= 5
aprovados = notas[mascara_nota5]
suspendidos = (~mascara_nota5).sum()
print(aprovados)
print(suspendidos)
```

**Variante extra:** Añade una segunda condición: extrae solo los sobresalientes (nota >= 9).

### Ejercicio 3 — Stock de productos agotados

**Enunciado:**  
Un almacén registra el stock de 10 productos.  
Obtén los índices de los productos agotados (stock = 0) para saber qué reabastecer.

**Ten en cuenta:**
- `np.where(condición)` devuelve una tupla con los **índices** donde la condición es `True`.
- `array == valor` crea una máscara de igualdad exacta.


```python
stock = np.array([120, 0, 45, 0, 300, 0, 18, 0, 77, 0])
mascara_0 = stock == 0

indices_agotados = np.where(mascara_0)[0]
print(indices_agotados)
```

**Variante extra:** Identifica también los productos con stock crítico (mayor que 0 pero menor que 20).

### Ejercicio 4 — Ventas por debajo del objetivo

**Enunciado:**  
Un equipo comercial tiene registradas las ventas mensuales en miles de euros.  
El objetivo mensual es 50 000 EUR. Devuelve los valores que **no** alcanzaron el objetivo y calcula el déficit total.

**Ten en cuenta:**
- El déficit de cada mes fallido es `objetivo - venta`.
- Puedes hacer operaciones aritméticas sobre valores filtrados.


```python
ventas = np.array([48.2, 63.5, 41.0, 55.8, 38.9, 70.2, 49.5, 52.1, 44.7, 60.0, 37.3, 58.6])
objetivo = 50.0 

noalcanzan = ventas < objetivo
ventas_fallidas = ventas[noalcanzan]

deficit = objetivo - ventas

print(ventas_fallidas,deficit.sum())
```

### Ejercicio 9 — Censurar valores extremos (clipping manual)

**Enunciado:**  
Un sensor de temperatura tiene lecturas erróneas: valores por debajo de -20°C o por encima de 60°C son imposibles.  
Reemplaza esos valores con `np.nan` para marcarlos como no válidos.

**Ten en cuenta:**
- `array[mascara] = valor` reemplaza **in place** (modifica el array original).
- `np.nan` solo funciona con arrays de tipo `float`, no `int`.
- Es buena práctica trabajar sobre una copia: `array.copy()`.


```python
temps_sensor = np.array([-25.0, 18.5, 62.3, 30.0, -22.1, 45.6, 15.0,
                          70.0, 28.9, -30.0, 50.2, 55.5], dtype=float)
temps_limpio = temps_sensor.copy()
fuera_rango = (temps_sensor < -20) | (temps_sensor > 60)

print("Lecturas erroneas=",fuera_rango.sum())
temps_limpio[fuera_rango] = np.nan
print(temps_limpio)
```

**Variante extra:** En lugar de NaN, reemplaza los valores erróneos con la media de los valores válidos.

### Ejercicio 10 — Sistema de calificaciones: convertir a letra

**Enunciado:**  
Convierte notas numéricas a letras usando reemplazos con máscaras:
- **F:** < 5 | **D:** 5-6.4 | **C:** 6.5-7.4 | **B:** 7.5-8.4 | **A:** 8.5-10

**Ten en cuenta:**
- Crea un array de strings: `np.full(shape, '?', dtype='U1')`.
- Aplica las máscaras en cualquier orden (son mutuamente excluyentes).


```python
puntos = np.array([3.2, 7.8, 5.5, 9.1, 6.5, 8.5, 4.9, 7.0, 8.0, 10.0, 6.0, 9.8])

letras = np.full(puntos.shape, '?',dtype='U1')
mascara_f = puntos < 5
mascara_d = (puntos >= 5) & (puntos <= 6.4)
mascara_c = (puntos >= 6.5) & (puntos <= 7.4)
mascara_b = (puntos >= 7.5) & (puntos <= 8.4)
mascara_a = (puntos >= 8.5) & (puntos <= 10)

letras[mascara_f] = "F"
letras[mascara_d] = "D"
letras[mascara_c] = "C"
letras[mascara_b] = "B"
letras[mascara_a] = "A"

print(letras)
```

**Variante extra:** Añade 'matricula de honor': nota >= 9.5 -> letra 'MH' (necesitarás `dtype='U2'`).
