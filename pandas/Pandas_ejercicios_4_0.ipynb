```python
import pandas as pd
import numpy as np

# Crear DataFrame
df = pd.DataFrame({
    'A': [1, 2, np.nan],
    'B': [np.nan, 5, np.nan],
    'C': [7, 8, 9]
})

# Añadir columna completamente nula
df['D'] = np.nan

# Mostrar DataFrame original
print("DataFrame original:")
print(df)

# 1. Eliminando filas con algún NaN
print("\n1. Filas con algún NaN eliminadas:")
print(df.dropna())

# 2. Eliminando columnas con algún NaN
print("\n2. Columnas con algún NaN eliminadas:")
print(df.dropna(axis=1))

# 3. Eliminando solo filas completamente nulas
print("\n3. Filas completamente nulas eliminadas:")
print(df.dropna(how='all'))

# 4. Eliminando solo columnas completamente nulas
print("\n4. Columnas completamente nulas eliminadas:")
print(df.dropna(axis=1, how='all'))
```

```python
"""Dado el siguiente DataFrame sucio, realiza un pipeline de limpieza completo:
1. Detecta y cuenta los NaN por columna.
2. elimina filas donde TODAS las columnas de datos son NaN
3. rellena los NaN restantes de 'precio' con la mediana y los de 'categoria' con 'Desconocido'
4. verifica que no quedan nulos
"""
    
df = pd.DataFrame({
 'id': [1, 2, 3, 4, 5, 6],
 'precio': [10.5, np.nan, 8.0, np.nan, 15.0, np.nan],
 'categoria': ['A', None, 'B', None, 'A', None],
 'cantidad': [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
})

"1. Detecta y cuenta los NaN por columna."
df.isnull().sum()

"2. elimina filas donde TODAS las columnas de datos son NaN"
df = df.dropna(how='all')

"3. rellena los NaN restantes de 'precio' con la mediana y los de 'categoria' con 'Desconocido'"
df['precio'] = df['precio'].fillna(df['precio'].median())
df['categoria'] = df['categoria'].fillna('Desconocido')
print(df)

"4. verifica que no quedan nulos"
print(df.isnull().sum())
```
