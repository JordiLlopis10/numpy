## 1. El objeto `Series` de Pandas

Una `Series` de Pandas es un **array unidimensional de datos indexados**. 

Por defecto, es el clásico array que empieza en 0.

Se puede crear a partir de una lista o array de la siguiente manera:

````python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
data
````

Como se observa en la salida, la `Series` contiene tanto una secuencia de **valores** como una secuencia de **índices**, accesibles mediante los atributos `.values` e `.index`.

Los `.values` son simplemente un array de NumPy:

````python
data.values
````

El `.index` es un objeto similar a un array de tipo `pd.Index`, que veremos con más detalle luego.

````python
data.index
````

Al igual que con un array de NumPy, se puede acceder a los datos mediante el índice usando la notación de corchetes de Python:

````python
# Acceso a un elemento individual
data[1]
````

````python
# Acceso a varios datos mediante slicing
data[1:3]
````

Como veremos ahora, la `Series` de Pandas es mucho más general y flexible que el array unidimensional de NumPy que emula.

---

### `Series` como array generalizado de NumPy

La diferencia esencial entre una `Series` y un array de NumPy reside en el **índice**:

- Un **array NumPy** tiene un índice entero definido *implícitamente* por la librería.
- Una **`Series` de Pandas** tiene un índice *explícitamente definido* por el programador y asociado a los valores.

<div class="alert alert-block alert-info">
<b>Pregunta:</b> ¿Qué diferencia hay entre implícito y explícito?
</div>

Esta definición explícita del índice otorga a la `Series` funcionalidades adicionales. 

Por ejemplo, el índice no tiene por qué ser entero; puede ser de cualquier tipo. 

Aquí os dejo un ejemplo para usar cadenas de texto como índice:

````python
data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=['a', 'b', 'c', 'd'])
data
````

````python
# El acceso a los elementos funciona como se espera
data['b']
````

Incluso podemos usar índices no contiguos o no secuenciales:

````python
data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=[2, 5, 3, 7])
data
````

````python
data[5]
````

### `Series` como mejora de diccionario

Podemos pensar en una estructura `Series` de Pandas como una mejora de un diccionario Python:

- Un **diccionario** mapea claves arbitrarias a valores arbitrarios.
- Una **`Series`** mapea claves *tipadas* a valores *tipados*.

<div class="alert alert-block alert-success">
<b>Arbitrario:</b> <br>
Arbitrario significa que no hay restricción sobre el tipo. En un diccionario Python normal, cada clave y cada valor pueden ser de un tipo diferente sin ningún problema.
</div>

<div class="alert alert-block alert-success">
<b>Tipado:</b> <br>
Tipado significa que todos los elementos comparten el mismo tipo.
</div>


Este tipado es importante: igual que el array NumPy lo hace más eficiente que una lista Python, la información de tipo de una estructura `Series` la hace mucho más eficiente que un diccionario Python para ciertas operaciones.

La analogía entre `Series` y diccionario queda aún más clara al construir una `Series` directamente desde un diccionario Python:

````python
# Diccionario básico de python
persona = {
    "nombre": "Ana",
    "edad": 30,
    "ciudad": "Madrid",
    "activo": True
}

# Diccionario con claves (string) y valores del mismo tipo (enteros)

poblacion_dict = {
    'California': 38_332_521,
    'Texas':      26_448_193,
    'New York':   19_651_127,
    'Florida':    19_552_860,
    'Illinois':   12_882_135
}

# Lo convertimos a Series
poblacion = pd.Series(poblacion_dict)
poblacion
````

Por defecto, la `Series` se crea con el índice extraído de las **claves ordenadas** del diccionario. A partir de aquí, se puede realizar el acceso típico de diccionario:

````python
poblacion['California']
````

A diferencia de un diccionario, la `Series` también admite operaciones de array como el *slicing*:

````python
poblacion['California':'Illinois']
````

> ⚠️ **Nota:** El slicing con etiquetas en Pandas es **inclusivo** en ambos extremos (a diferencia del slicing estándar de Python que excluye el extremo derecho).

### Construcción de objetos `Series`

Ya hemos visto varias formas de construir una `Series`; todas son variantes de:

```python
pd.Series(data, index=index)
```

donde `index` es un argumento opcional y `data` puede ser de muchos tipos.

**A partir de una lista o array NumPy** → el índice por defecto es una secuencia entera:

````python
pd.Series([2, 4, 6])
````

**A partir de un número** → el valor se repite para llenar el índice especificado:

````python
pd.Series(5, index=[100, 200, 300])
````

**A partir de un diccionario** → el índice por defecto son las claves del diccionario respetando el orden:

````python
pd.Series({2: 'a', 1: 'b', 3: 'c'})
````

En cada caso, el índice puede establecerse explícitamente si se prefiere un resultado diferente. Observa que la `Series` solo incluye las claves indicadas explícitamente:

````python
pd.Series({2: 'a', 1: 'b', 3: 'c'}, index=[3, 2])
````
