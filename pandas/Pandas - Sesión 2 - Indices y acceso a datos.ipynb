## Selección de Datos en Series

Como vimos en la sesión anterior, un objeto `Series` se comporta en muchos aspectos como un array NumPy unidimensional y, en otros, como un diccionario estándar de Python. Tener en mente estas dos analogías nos ayuda a comprender los patrones de indexación y selección de datos.

### Series como diccionario

Al igual que un diccionario, el objeto `Series` establece una correspondencia entre un conjunto de claves y un conjunto de valores:

```python
import pandas as pd

# Crear una Serie con índice de letras
datos = pd.Series([0.25, 0.5, 0.75, 1.0],
                  index=['a', 'b', 'c', 'd'])
datos
```

```python
# Acceso a un elemento por su índice (igual que en un diccionario)
datos['b']
```


También podemos usar expresiones y métodos típicos de diccionarios para examinar claves/índices y valores:

```python
# Comprobar si una clave existe en la Serie
'a' in datos
```

```python
# Obtener las claves (índices) de la Serie
datos.keys()
```

```python
# Obtener pares (clave, valor) como lista de tuplas
list(datos.items())
```


Los objetos `Series` también pueden modificarse con sintaxis similar a la de los diccionarios. Del mismo modo que se puede extender un diccionario asignando a una nueva clave, se puede extender una `Series` asignando a un nuevo valor de índice:

```python
# Agregar un nuevo elemento a la Serie
datos['e'] = 1.25
datos
```

### Series como array unidimensional

Una `Series` amplía la interfaz tipo diccionario y proporciona selección de elementos al estilo array mediante los mismos mecanismos básicos que NumPy: *slicing*, *enmascaramiento* e *indexación múltiple*.

```python
# Slicing por índice explícito (el índice final se INCLUYE)
datos['a':'c']
```

```python
# Slicing por índice implícito entero (el índice final se EXCLUYE)
datos[0:2]
```

```python
# Máscaras booleanas: seleccionar valores dentro de un rango
datos[(datos > 0.3) & (datos < 0.8)]
```

```python
# Indexación múltiple: seleccionar múltiples elementos por nombre
datos[['a', 'e']]
```



> ⚠️ **Nota importante sobre el slicing:** al usar índice explícito (`datos['a':'c']`), el índice final **se incluye**; al usar índice implícito entero (`datos[0:2]`), el índice final **se excluye**.

### Indexadores: loc e iloc

Estas convenciones pueden generar confusión. Por ejemplo, si una `Series` tiene un índice entero explícito, una operación de indexación como `datos[1]` usará el índice explícito, mientras que un slicing como `datos[1:3]` usará el índice implícito de Python.

```python
# Serie con índice entero explícito (no empieza en 0)
datos = pd.Series(['a', 'b', 'c'], index=[1, 3, 5])
datos
```

```python
# Al indexar con un entero, se usa el índice EXPLÍCITO (= definido por el programador)
datos[1]
```

**Salida esperada:** `'a'`

```python
# Al hacer slicing, se usa el índice IMPLÍCITO (= definido por Python)
datos[1:3]
```

Para resolver esta ambigüedad, Pandas ofrece el método loc.

**`loc`** — siempre referencia el índice **explícito**:

```python
# loc usa el índice explícito (=programador)
datos.loc[1]
```

```python
# Slicing con loc: incluye el límite final
datos.loc[1:3]
```

**`iloc`** — siempre referencia el índice **implícito** (definido por Python):

```python
# iloc usa la posición implícita (=python)
datos.iloc[1]
```

```python
# Slicing con iloc: excluye el límite final (comportamiento estándar de Python)
datos.iloc[1:3]
```



> 💡 **Principio guía:** En Python, *"explícito es mejor que implícito"*. Usar `loc` e `iloc` hace el código más legible y evita bugs sutiles, especialmente con índices enteros. **Se recomienda usar siempre estos indexadores.**

## Selección de Datos en DataFrame

Recordemos que un `DataFrame` se comporta en muchos aspectos como un array bidimensional o estructurado, y en otros como un diccionario de objetos `Series` que comparten el mismo índice.

### DataFrame como diccionario

La primera analogía es pensar en el `DataFrame` como un diccionario de objetos `Series` relacionados. Retomemos el ejemplo de áreas y poblaciones de estados:

```python
# Crear Series de área y población por estado
area = pd.Series({'California': 423967, 'Texas': 695662,
                  'Nueva York': 141297, 'Florida': 170312,
                  'Illinois': 149995})

pob = pd.Series({'California': 38332521, 'Texas': 26448193,
                 'Nueva York': 19651127, 'Florida': 19552860,
                 'Illinois': 12882135})

# Construir un DataFrame combinando ambas Series
datos = pd.DataFrame({'area': area, 'pob': pob})
datos
```

Se puede acceder a cada `Series` (columna) mediante indexación de diccionario:

```python
# Acceso a columna por nombre (estilo diccionario)
datos['area']
```



Equivalentemente, se puede acceder mediante atributo cuando el nombre de la columna es una cadena:

```python
# Acceso a columna mediante atributo (notación con punto)
datos.area
```

Al igual que con `Series`, esta sintaxis tipo diccionario sirve también para añadir nuevas columnas:

```python
# Añadir columna de densidad de población (hab/km²)
datos['densidad'] = datos['pob'] / datos['area']
datos
```

### DataFrame como array bidimensional

También podemos ver el `DataFrame` como un array bidimensional mejorado. Podemos inspeccionarlo con el atributo `values`:

```python
# Ver los datos brutos como array NumPy
datos.values
```

```python
# Transponer el DataFrame: intercambiar filas y columnas
datos.T
```

Con **`iloc`** podemos acceder a las filas y las columnas:

```python
# iloc: primeras 3 filas, primeras 2 columnas (por posición)
datos.iloc[:3, :2]
```

Con **`loc`** indexamos por etiquetas explícitas:

```python
# loc: filas hasta 'Illinois' (inclusive), columnas hasta 'pob' (inclusive)
datos.loc[:'Illinois', :'pob']
```

Dentro de estos indexadores podemos combinar máscaras booleanas e indexación múltiple:

```python
# Seleccionar filas donde la densidad > 100, mostrando solo columnas 'pob' y 'densidad'
datos.loc[datos.densidad > 100, ['pob', 'densidad']]
```



Estas convenciones de indexación también sirven para **modificar** valores:

```python
# Modificar un valor concreto: fila 0, columna 2 (densidad de California)
datos.iloc[0, 2] = 90
datos
```

También podemos usar máscaras booleanas como haciamos con NumPy:

```python
# Enmascaramiento directo: también opera sobre FILAS
datos[datos.densidad > 100]
```


---

## Resumen

| Operación | Sintaxis | Resultado |
|---|---|---|
| Acceder a columna | `df['col']` o `df.col` | Serie |
| Acceder a fila por etiqueta | `df.loc['fila']` | Serie |
| Acceder a fila por posición | `df.iloc[n]` | Serie |
| Slicing de filas por etiqueta | `df.loc['a':'b']` | DataFrame |
| Slicing de filas por posición | `df.iloc[n:m]` | DataFrame |
| Slicing directo | `df[n:m]` | DataFrame (filas) |
| Enmascaramiento directo | `df[mask]` | DataFrame (filas) |
| Selección 2D con etiquetas | `df.loc[filas, cols]` | DataFrame |
| Selección 2D con posiciones | `df.iloc[filas, cols]` | DataFrame |
