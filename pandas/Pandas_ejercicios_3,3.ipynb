```python
superficie = pd.Series({'España': 505900, 'Francia': 551700,
'Alemania': 357100, 'Italia': 301300})
poblacion = pd.Series({'España': 47_400_000, 'Francia': 68_000_000,
'Alemania': 84_400_000, 'Portugal': 10_300_000})
densidad = (poblacion / superficie).round(2)
print(densidad)
```

```python
df1 = pd.DataFrame({'X': [1, 2], 'Y': [3, 4]},
index=['fila1', 'fila2'])
df2 = pd.DataFrame({'Y': [10, 20, 30], 'Z': [40, 50, 60], 'X': [5, 6, 7]},
index=['fila1', 'fila2', 'fila3'])

df1 + df2
```

```python
print(df1.add(df2, fill_value=0))
```

```python
media = df1.stack().mean()
df1.add(df2, fill_value=media)
```

```python
ventas = pd.DataFrame(
{'Tienda_A': [100, 200, 150], 'Tienda_B': [80, 300, 120]},
index=['Producto1', 'Producto2', 'Producto3']
)
precio = pd.Series({'Producto1': 5.0, 'Producto2': 3.5, 'Producto3': 8.0})
ventas.mul(precio, axis=0)
```

```python
tienda_c = pd.Series([50, 400], index=["Producto1", "Producto2"])
ventas["Tienda_C"] = tienda_c
ventas = ventas.fillna(0)
ventas.mul(precio, axis=0)
```
