```python
import pandas as pd
import numpy as np
```

```python

arr = np.array([3, np.nan, 7, np.nan, 12, 1])

# Cálculos con funciones estándar
media_std = np.mean(arr)
desviacion_std = np.std(arr)
mediana_std = np.median(arr)

# Cálculos con funciones que ignoran NaN
media_nan = np.nanmean(arr)
desviacion_nan = np.nanstd(arr)
mediana_nan = np.nanmedian(arr)
print("Resultados con funciones estándar:")
print(f"Media: {media_std}, Desviación Típica: {desviacion_std}, Mediana: {mediana_std}")
print("\nResultados con funciones que ignoran NaN:")
print(f"Media: {media_nan}, Desviación Típica: {desviacion_nan}, Mediana: {mediana_nan}")

```

```python
df = pd.DataFrame({
 'nombre': ['Ana', 'Luis', None, 'Eva', None],
 'edad': [25, None, 32, None, 28],
 'salario': [None, None, None, 4000, 3200],
 'ciudad': ['Madrid', 'BCN', 'Madrid', 'BCN', 'Madrid'],
})
# 1. Número de nulos por columna
nulos = df.isnull().sum()

# 2. Porcentaje de nulos por columna
porcentaje = (df.isnull().mean() * 100).sort_values(ascending=False)

# 3. Columnas con más de un 40% de nulos
mas_40 = porcentaje[porcentaje > 40]

print(nulos,"\n\r")
print(porcentaje,"\n\r")
print(mas_40)
```

```python
df = pd.DataFrame({
 'nombre': ['Ana', 'Luis', 'Marta', 'Pedro', 'Sara'],
 'email': ['ana@x.com', None, 'marta@x.com', None, 'sara@x.com'],
 'departamento': ['IT', 'HR', None, 'IT', 'HR'],
 'salario': [3000, 2500, np.nan, 4000, np.nan],
})

# Eliminar filas donde email o departamento sean nulos
df_filtrado = df.dropna(subset=['email', 'departamento'])

print("\nDespués de eliminar nulos en 'email' o 'departamento':")
print(df_filtrado)

# Eliminar también filas donde salario sea nulo
df_final = df_filtrado.dropna(subset=['salario'])

print("\nDespués de eliminar también nulos en 'salario':")
print(df_final)
```

```python
df = pd.DataFrame({
 'A': [1, np.nan, 3, np.nan, 5],
 'B': [np.nan, np.nan, np.nan, np.nan, np.nan],
 'C': [10, 20, np.nan, 40, 50],
 'D': [np.nan, 2, np.nan, 4, np.nan],
})
# Contar valores no nulos por columna
validos = df.notnull().sum()

print("\nValores válidos por columna:")
print(validos)

# Eliminar columnas con menos de 3 valores no nulos
df_filtrado = df.dropna(axis=1, thresh=3)

print("\nDataFrame después de aplicar dropna(axis=1, thresh=3):")
print(df_filtrado)
```

```python
df = pd.DataFrame({
 'dia': ['Lun', 'Mar', 'Mie', 'Jue', 'Vie'],
 'temperatura': [22.0, np.nan, 18.0, np.nan, 25.0],
 'condicion': ['Lluvia', None, 'Nublado', None, 'Soleado'],
 'viento': [15, np.nan, 10, np.nan, 20],
})
print(df)
# Calcular la media de temperatura
media_temp = df['temperatura'].mean()

# Rellenar NaN con un solo fillna()
df.fillna({
    'temperatura': media_temp,
    'condicion': 'Despejado',
    'viento': 0
}, inplace=True)

print("\nDataFrame sin nulos:")
print(df)

# Verificar que no quedan nulos
print("\n¿No quedan nulos?:", df.isnull().sum().sum() == 0)
```

```python
df = pd.DataFrame({
    'paciente_id': [101, 102, None, 104, 105, None],
    'edad': [45, np.nan, 60, np.nan, 33, 70],
    'diagnostico': ['Gripe', None, 'Asma', 'Gripe', None, 'Asma'],
    'notas': [None, None, None, None, 'Leve', None],
})

print("DataFrame original:")
print(df)

# 1. Porcentaje de nulos por columna
print("\nPorcentaje de nulos por columna:")
porcentaje_nulos = df.isnull().mean() * 100
print(porcentaje_nulos)

# 2. Eliminar columna con más del 60% de nulos
columnas_eliminar = porcentaje_nulos[porcentaje_nulos > 60].index
df = df.drop(columns=columnas_eliminar)

# 3. Eliminar filas donde 'paciente_id' es nulo
df = df.dropna(subset=['paciente_id'])

# 4. Rellenar 'edad' con la mediana y 'diagnostico' con 'Sin registro'
mediana_edad = df['edad'].median()

df = df.fillna({
    'edad': mediana_edad,
    'diagnostico': 'Sin registro'
})

# 5. Verificar resultado
print("\nDataFrame limpio:")
print(df)

print("\nNulos restantes por columna:")
print(df.isnull().sum())

print("\n¿No quedan nulos?:", df.isnull().sum().sum() == 0)
```
